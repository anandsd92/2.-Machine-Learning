```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Pandas - Pandas is a powerful framework to work with structured data like CSV, Tables, and Data base
try:
    dataset = pd.read_csv("insurance_pre.csv") # Method to import csv data into 'pd' object
    print("Data loaded successfully")
    print(f"Rows: {len(dataset)}, Columns: {len(dataset.columns)}")
except Exception as e:
    print(f" Either file not found or Unable to open or Unexpected error: {e}")
```

```python
dataset.columns
```

```python
#In Pandas, the get_dummies() function converts categorical variables into dummy/indicator variables (known as one-hot encoding)
#Methods avilable to this 1. One Hot Encoding 2. Using Sckit learn Label Encoder 3. Direct Lable Encoding
dataset=pd.get_dummies(dataset, columns=['sex', 'smoker'], drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
indepent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indepent, dependent, test_size = 1/3, random_state = 0)
```

```python
#Importing Linear Regression from scikit-learns linear_model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() # This object holds all the model information
#fit() method is used to train the model
regressor.fit(X_train, y_train) # Passing training input value and output value to the model
```

```python
#Viewing the weight and bias of the model
weight = regressor.coef_
bias=regressor.intercept_
print("Weight of the model={}".format(weight))
print("Orgin of the model, nothing has been done what will be the profit={}".format(bias))
```

```python
#Pass the X_test for the prediction model
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
rscore=r2_score(y_test,y_pred)
```
