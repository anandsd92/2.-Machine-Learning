```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
#Pandas - Pandas is a powerful framework to work with structured data like CSV, Tables, and Data base
#Pandas - Pandas is a powerful framework to work with structured data like CSV, Tables, and Data base
try:
    dataset = pd.read_csv("insurance_pre.csv") # Method to import csv data into 'pd' object
    print("Data loaded successfully")
    print(f"Rows: {len(dataset)}, Columns: {len(dataset.columns)}")
except Exception as e:
    print(f" Either file not found or Unable to open or Unexpected error: {e}")
```

```python
#The head() method in pandas is used to quickly preview the first few rows of a DataFrame
dataset.head
```

```python
#The dataframe.describe() method in pandas is used to generate descriptive statistics that summarize the central tendency, dispersion, and 
#shape of a dataset's distribution
dataset.describe()
```

```python
# Data distribution
plt.title('Salary Distribution Plot')
sns.histplot(dataset['charges'])
plt.show()
```

```python
# Relationship between Salary and Level
plt.scatter(dataset['age'], dataset['charges'], color = 'lightcoral')
plt.title('Age vs Insurance Charge')
plt.xlabel('age')
plt.ylabel('charges')
plt.box(False)
plt.show()
```

```python
#In Pandas, the get_dummies() function converts categorical variables into dummy/indicator variables (known as one-hot encoding)
#Methods avilable to this 1. One Hot Encoding 2. Using Sckit learn Label Encoder 3. Direct Lable Encoding
dataset=pd.get_dummies(dataset, columns=['sex', 'smoker'], drop_first=True)
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)
```
